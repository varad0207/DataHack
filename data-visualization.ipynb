# Data Visualization

Data visualization plays a crucial role in exploring datasets and its vital for understanding the underlying patterns, relationships and insights within the data

## Tools i used to perform data visualization:
1. Matplotlib - Python plotting library, used to create range of plots
2. Seaborn - Python library built on top of Matplotlib, used to create variety of complex plots like, heatmaps, countplot, pairplot, histogram, distribution plot

## Libraries used: 
1. Numpy - for numerical computing
2. Pandas - for it's core data structre, DataFrame, a 2D table that organizes data into rows and columns making it easir for data manipulation and analysis
3. Matplotlib
4. Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Countplot

```python
plt.figure(figsize=(6,3),dpi=100)
sns.countplot(x=df['target'])
```

In above countplot can see that in our dataset there is an imbalance of classes, i.e. the data is more biased toward one class then another

## Heatmap

```python
plt.figure(figsize=(22,11))
sns.heatmap(df.corr(),annot=True)
```

Due to large size of dataset, visualizing the data becomes a complex task.
The heatmap is generally used to find correlations between the variables
