# Predicting Target value using Logisitic Regression Model

```python
# importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from joblib import dump,load
```

```python
model = load('logistic_reg_model.joblib')
```

```python
df = pd.read_csv('test.csv')
```

```python
df
```

```python
df.set_index('ID_code')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X = df.drop('ID_code',axis=1)
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=200)
```

```python
pca_X = pca.fit_transform(scaled_X)
```

```python
pca_X.shape
```

```python
model_pred = model.predict(pca_X)
```

```python
np.count_nonzero(model_pred==1)
```

```python
df['target'] = model_pred
```

```python
df
```

```python
df.to_csv('predicted.csv')
```
